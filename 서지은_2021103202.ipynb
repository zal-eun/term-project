# 주제 : 제주도 관광객들의 선호 관광지, 여행 평가 분석을 통한 관광 서비스질 개선 


## 1. 주제 선정 이유 및 가설
 제주도는 한국에 유명한 관광 도시로, 매년 국내외 수많은 관광객들이 찾고 있다. 제주도를 찾는 이들이 어떤 종류의 관광지를 선호하며, 어디에 위치해있는지 확인해보고자 한다. 이를 통해서 앞으로 제주도 관광 산업에 있어서의 발전 사항을 연구해보고 관광 산업의 경제적 발전을 도모하고자 한다. 또한 보다 더 많은 관광객들의 편리한 관광을 위해, 앞서 알아본 결과를 바탕으로, 관광 서비스질 개선을 위한 방안을 제안하고자 한다.  
 
 제시한 주제에 대해 작성한 가설은 다음과 같다. 
 
 (1) 국내외 관광객들이 여행지를 제주도로 정하는 이유는 자연경관 감상에 목적이 있을 것이다.  
 
 (2) 국내외 관광객들은 제주도 대중교통 이용에 있어 불편함을 느낄 것이다.

  따라서 국내외 관광객들은 자연경관 감상을 목적으로 제주도를 여행 목적지로 설정하는 경우가 많으며 이때 대중교통 이용에 있어 불편함을 느낀다는 가설을 데이터 분석을 통해 입증하고자 한다.

### <1> 내국인 제주여행 고려요인(2014-2019)

#####

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df1 = pd.read_csv("C:\\Users\\82105\\Desktop\\파이썬 텀프 자료 모음\\획득데이터\\통계청 출처 CSV\\내국인_제주여행_고려요인_20221122003749.csv", encoding='euc-kr')
df1= df1.iloc[1:3, 2:13]
df1 = df1.transpose()
df1.columns =['설문문항','ratio']
df1['ratio'] = df1['ratio'].str.replace(',','').astype(float)
df1[1::].plot.bar(x='설문문항', y='ratio')
df1
```

### 2) 제주특별자치도 교통정보센터, Open Api, CITS-검색어 정보
<br>
<br>

**발급받은 key 값을 통해 open api를 불러오는 코드를 작성하였다.**

```python
import requests
import pandas as pd
url2 = 'http://api.jejuits.go.kr/api/infoSearchList?code=860528&startDt=20210101&endDt=20210131&dateType=M&queryType=1'
print(url2)
result_dict2 = requests.get(url2).json()
result_dict2
```

<br>


**불러낸 API 자료를 분석이 용이하도록 "colums"를 붙여 정렬한다. <br>검색 횟수가 많은 항목만 선별하기 위하여 인덱스 번호 100을 기준으로 임의로 슬라이싱 한다.**

```python
data_list2 = result_dict2['info']
cits_df = pd.DataFrame(data_list2)
columns=['sdate','search','cnt' ]
cits_df =pd.DataFrame(data= data_list2, columns=columns)
cits_df.head()
cits_df=cits_df.iloc[:101]
cits_df
```

<br>
<br>

**원활한 데이터 분석을 위해 다운받은 API 정보를 Excel 파일로 변환, 저장한다.**

```python
cits_df.to_excel("C:\\Users\\82105\\Desktop\\파이썬 텀프 자료 모음\\제주교통센터 출처 open api 변환\\cits.xlsx", index = False)
```

<br>
<br>

**정렬한 장소 목록 중, 자연경관과 관련있는 장소를 선별하기 위한 검열과정을 거친다.**

```python
cits1_df= cits_df[cits_df['search'].str.contains("해안도로")]
cits2_df= cits_df[cits_df['search'].str.contains("해수욕장")]
cits3_df= cits_df[cits_df['search'].str.contains("해안")]
cits4_df= cits_df[cits_df['search'].str.contains("산")]
cits5_df= cits_df[cits_df['search'].str.contains("굴")]
cits6_df= cits_df[cits_df['search'].str.contains("해변")]
cits7_df= cits_df[cits_df['search'].str.contains("나무")]
cits8_df= cits_df[cits_df['search'].str.contains("공원")]
cits9_df= cits_df[cits_df['search'].str.contains("수목원")]
cits10_df= cits_df[cits_df['search'].str.contains("외돌개")]
cits11_df= cits_df[cits_df['search'].str.contains("숲길")]
cits12_df= cits_df[cits_df['search'].str.contains("주상절리")]
cits13_df= cits_df[cits_df['search'].str.contains("오름")]
cits14_df= cits_df[cits_df['search'].str.contains("폭포")]


cits15_df= cits_df[cits_df['search'].str.contains("성산일출봉")]
cits16_df= cits_df[cits_df['search'].str.contains("섭지코지")]
```

<br>
<br>

**선별한 자료들 합쳐서 새로운 데이터 프레임으로 저장한다.**

```python
citsmerge = pd.concat([cits1_df,cits2_df,cits3_df,cits4_df,cits5_df,cits6_df,cits7_df,cits8_df,cits9_df,cits10_df,cits11_df,cits12_df,cits13_df,cits14_df,cits15_df,cits16_df])
citsmerge
citsmerge_df = pd.DataFrame(citsmerge)
```

<br>

**두 데이터의 칼럼명을 맞추기 위해 'search'을 'title'로 바꾼다.**

```python
citsmerge_df.rename(columns={'search':'title'},inplace = True)
```

<br>
<br>

**병합한 데이터 프레임을 Excel 파일로 변환, 저장한다.** 

```python
citsmerge_df.to_excel("C:\\Users\\82105\\Desktop\\파이썬 텀프 자료 모음\\가공데이터\\cits2.xlsx", index = False)
citsmerge_df
```

### 3) 관광지 데이터 지도에 표시
<br>
<br>

**선별한 관광지 위치를 가시적으로 보기 위하여 지도에 표시한다. 한라산을 기준으로 제주도가 한번에 보이는 크기로 지도가 구현될 수 있도록 설정한다. <br>
모든 관광지가 표시될 수 있도록 반복 횟수를 설정한다. 원 사이즈는 병합한 정보 중 검색어 횟수에 비례하도록 설정했다.**

```python
import folium

Mt_Hanla=[33.362500,126.533694]
map_jeju = folium.Map(location = Mt_Hanla, zoom_start =10)

for i in range (len(incor)):
    name = incor['title'][i]
    count= incor['cnt'][i]
    size =int(incor['cnt'][i])*0.1
    lat=float(incor['latitude'][i])
    long = float(incor['longitude'][i])    
    
    folium.CircleMarker((lat,long),radius= size, color='red',popup=name).add_to(map_jeju)
    

map_jeju
```
